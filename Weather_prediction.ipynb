```python
import pandas as pd
```

```python
df=pd.read_csv("/seattle-weather.csv")     #importing the dataset
```

```python
df.head()
```

```python
df.isna().sum()                           #checking for null values
```

```python
X=df.drop(columns=["date","weather"])     #defining features(X) and traget(y)
y=df["weather"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)       #splitting the dataset into train and test data
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()                               #defining the model
```

```python
model.fit(X_train,y_train)                                #fitting the model
```

```python
prediction=model.predict(X_test)                         #predicting the test data
```

```python
prediction
```

```python
from sklearn.metrics import classification_report    #importing the classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
test_weather={"precipitation":10,"temp_max":20,"temp_min":10,"wind":5.4}     #defining the test data
test_df=pd.DataFrame([test_weather])
test_df
```

```python
model.predict(test_df)                                                      #testing the model on new test data
```
